```python
# Musterlösung

alter_array = np.array([14, 16, 32, 25, 81])
```

*Wir wollen nun das durchschnittliche Alter von diesen Personen ausrechnen. Dazu können wir entweder den Befehl `np.sum()` benutzen, um die Alter in dem Array zu summieren und dann das ganze durch die Anzahl an Leuten teilen (dazu kannst du `len()` benutzen, dies gibt die Länge des Arrays an), oder wir können gleich den Befehl `np.mean()` benutzen.
```python
# Musterlösung

# Methode 1
durchschnittliches_alter = np.sum(alter_array) / len(alter_array)
print('Das durchschnittliche Alter beträgt:', durchschnittliches_alter)

# Methode 2
durchschnittliches_alter = np.mean(alter_array)
print('Das durchschnittliche Alter beträgt:', durchschnittliches_alter)
```

**<span style="color: #EB6ECB;">Aufgabe 6.2:</span>** *Wir können Arrays aber auch verändern, indem wir Werte hinzufügen oder löschen. Ihr habt schon die Temperaturen von Januar bis Mai gesammelt, aber es fehlen noch die Daten für die anderen Monate.*

```python
temperaturliste = np.array([8.1, 8.7, 8.7, 11.6, 18.8]) # Temperaturen in °C von Januar bis Mai
```

*Man kann also sehen, dass die Durchschnittstemperatur im Mai 18,8 °C war. 
Jetzt ist es Ende Juni und ihr müsst die Durchschnittstemperatur für Juni hinzufügen. Wir wollen nicht jedes Mal, wenn wir einen neuen Monat messen, ein neues Array erstellen, sondern unsere bestehende *`Temperaturliste`* erweitern. Dazu kannst du die NumPy-Funktion *`np.append()`* benutzen, um einen neuen Eintrag an die Liste anzuhängen. Die Syntax von *`np.append()`* sieht so aus: *`np.append(array, wert)`*. Du sagst dieser NumPy-Funktion also, welches Array *`array`* du hast (deine Temperaturliste) und welches neue Element *`wert`* du hinzufügen möchtest (die Temperatur für den neuen Monat Juni) und die Funktion gibt dir dann ein neues Array mit den angehängten Werten aus!

Das bedeutet, dass die Temperatur im Januar den Index 0 hat, Februar den Index 1 usw. Die Indexierung bei Listen sollte euch bereits aus den letzten Wochen bekannt sein. Lösche nun mithilfe der Funktion *`np.delete()`* den Eintrag des Monats Januar.*

Die Struktur sieht so aus: *`np.delete(array, index)`*. Dabei ist

- *`array`* eure Liste oder euer Array, aus dem ihr etwas entfernen möchtet,
- *`index`* die Position des Wertes in eurem Array, das ihr entfernen möchtet.

### <span style="color: #C26BCA;"> Indizierung von Arrays </span>

#### <span style="color: #C26BCA;"> Mehrere Linien in einem Diagramm </span>
Manchmal möchten wir mehrere Linien in einem Diagramm darstellen, um Daten zu vergleichen. Zum Beispiel könnten wir uns die minimale und die maximale Temperatur in Rom anzeigen lassen. Dazu können wir einfach den Befehl `plt.plot()` mehrfach wiederholen.

```python
x = range(1, 13) # Monate von 1 bis 12
maximale_temperatur          = [13.1, 14.3, 17.1, 19.2, 24.4, 28.7, 31.8, 32.1, 27.1, 22.8, 17.2, 13.3]  # Temperaturen in °C
durchschnittliche_temperatur = [8.1, 8.7, 8.7, 11.6, 18.8, 22.8, 25.4, 25.7, 21.4, 17.6, 12.6, 8.9]  # Temperaturen in °C
minimale_temperatur          = [3, 3, 6.1, 8.2, 13.1, 16.8, 19, 19.3, 15.6, 12.3, 7.9, 4.4]  # Temperaturen in °C

plt.figure(figsize=(14,9))

plt.plot(x, maximale_temperatur)
plt.plot(x, durchschnittliche_temperatur)
plt.plot(x, minimale_temperatur)

plt.xlabel('Monat', fontsize=25)
plt.ylabel('Temperatur T[°C]', fontsize=25)
plt.title('Temperatur in Rom', fontsize=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.show()
```

Warte... Welche Linie ist jetzt welche? Dafür können wir eine Legende verwenden! Der Befehl `plt.plot()` hat das Argument `label` welches uns erlaubt eine Beschriftung für diese Linie hinzuzufügen. Am Ende kann dann noch mit `plt.legend()` eine Legende eingefügt werden, die uns erklärt wofür die ganzen Beschriftungen stehen:

```python
x = range(1, 13) # Monate von 1 bis 12
maximale_temperatur          = [13.1, 14.3, 17.1, 19.2, 24.4, 28.7, 31.8, 32.1, 27.1, 22.8, 17.2, 13.3]  # Temperaturen in °C
durchschnittliche_temperatur = [8.1, 8.7, 8.7, 11.6, 18.8, 22.8, 25.4, 25.7, 21.4, 17.6, 12.6, 8.9]  # Temperaturen in °C
minimale_temperatur          = [3, 3, 6.1, 8.2, 13.1, 16.8, 19, 19.3, 15.6, 12.3, 7.9, 4.4]  # Temperaturen in °C

plt.figure(figsize=(14,9))

plt.plot(x, maximale_temperatur, label='Maximale Temperatur')
plt.plot(x, durchschnittliche_temperatur, label='Durchschnittliche Temperatur')
plt.plot(x, minimale_temperatur, label='Minimale Temperatur')

plt.legend(fontsize=15)

plt.xlabel('Monat', fontsize=25)
plt.ylabel('Temperatur T[°C]', fontsize=25)
plt.title('Temperatur in Rom', fontsize=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.show()
```

Wir benutzen hier gleich wieder das Argument `fontsize` um die Größe des Textes anzupassen.

Als letztes wollen wir jetzt noch die Farben und den Linienstil anpassen. Das können wir auch ganz einfach über die Argumente `color` für die Farbe und `linestyle` für den Linienstil regeln. Was steht uns aber zur Auswahl? Dazu schauen wir einfach in der Dokumentation nach:
- die Liste an verfügbaren Farben finder wir unter: https://matplotlib.org/stable/gallery/color/named_colors.html

Nutze nun NumPy, um den Mittelwert von jedem Array zu bestimmen und plotte den mittleren Durchmesser der Kristalle als Funktion der Zeit. Werden deine Kristalle größer oder bleiben sie ungefähr gleich groß?*

## <span style="color: #9C68C8;">Bonus: Größen verändern</span>


Um unsere Plots zugänglicher zu machen, können wir den Graphen auch noch ein bisschen vergrößern. Dafür benutzen wir den Befehl `plt.figure()`. Dieser erstellt ein leeres Bild. Das Ganze könnt ihr euch wie eine unbemalte Leinwand oder ein unbedrucktes Blatt Papier vorstellen. Da wir natürlich die Leinwand oder das Papier vor dem Malen oder Drucken brauchen, muss dieser Befehl auch vor allen anderen stehen. Die Größe kann dann mit dem Argument `figsize` angepasst werden. Die Größe wird dabei durch `(B,H)` dargestellt, wobei `B` die Breite und `H` die Höhe sind. Matplotlib füllt dann die gesamte Leinwand bis zum Rand aus! So sieht das dann aus:



```python
plt.figure(figsize=(14,9)) # Dieser Befehl steht immer vor allen Befehlen die die Visualisierung betreffen

plt.plot()
plt.xlabel('x-Achsen Beschriftung')
plt.ylabel('y-Achsen Beschriftung')
plt.title('Überschrift')

plt.show() # Dieser Befehl steht immer nach allen anderen Befehlen die Visualisierung betreffen
```

Das einzige Problem, das uns noch bleibt, ist, dass zwar das Bild größer ist, aber die Achsenbeschriftung nicht. Dies lässt sich durch das Argument `fontsize` (also Schriftgröße auf Englisch) beheben. 

```python
plt.figure(figsize=(14,9)) 

plt.plot()
plt.xlabel('x-Achsen Beschriftung', fontsize=25)
plt.ylabel('y-Achsen Beschriftung', fontsize=25)
plt.title('Überschrift', fontsize=30)

plt.show()
```

Dasselbe geht übrigens auch mit den Zahlen an den Achsen in dem man den parameter `fontsize` in `plt.xticks()` und `plt.yticks()` anpasst

```python
plt.figure(figsize=(14,9)) 

plt.plot()
plt.xlabel('x-Achsen Beschriftung', fontsize=25)
plt.ylabel('y-Achsen Beschriftung', fontsize=25)
plt.title('Überschrift', fontsize=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.show()
```
