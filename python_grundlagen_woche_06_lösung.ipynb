#### <span style="color: #C26BCA;"> Mehrere Linien in einem Diagramm </span>
Manchmal mchten wir mehrere Linien in einem Diagramm darstellen, um Daten zu vergleichen. Zum Beispiel knnten wir uns die minimale und die maximale Temperatur in Rom anzeigen lassen. Dazu knnen wir einfach den Befehl `plt.plot()` mehrfach wiederholen.

```python
x = range(1, 13) # Monate von 1 bis 12
maximale_temperatur          = [13.1, 14.3, 17.1, 19.2, 24.4, 28.7, 31.8, 32.1, 27.1, 22.8, 17.2, 13.3]  # Temperaturen in C
durchschnittliche_temperatur = [8.1, 8.7, 8.7, 11.6, 18.8, 22.8, 25.4, 25.7, 21.4, 17.6, 12.6, 8.9]  # Temperaturen in C
minimale_temperatur          = [3, 3, 6.1, 8.2, 13.1, 16.8, 19, 19.3, 15.6, 12.3, 7.9, 4.4]  # Temperaturen in C

plt.figure(figsize=(14,9))

plt.plot(x, maximale_temperatur)
plt.plot(x, durchschnittliche_temperatur)
plt.plot(x, minimale_temperatur)

plt.xlabel('Monat', fontsize=25)
plt.ylabel('Temperatur T[C]', fontsize=25)
plt.title('Temperatur in Rom', fontsize=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.show()
```

Warte... Welche Linie ist jetzt welche? Dafr knnen wir eine Legende verwenden! Der Befehl `plt.plot()` hat das Argument `label` welches uns erlaubt eine Beschriftung fr diese Linie hinzuzufgen. Am Ende kann dann noch mit `plt.legend()` eine Legende eingefgt werden, die uns erklrt wofr die ganzen Beschriftungen stehen:

```python
x = range(1, 13) # Monate von 1 bis 12
maximale_temperatur          = [13.1, 14.3, 17.1, 19.2, 24.4, 28.7, 31.8, 32.1, 27.1, 22.8, 17.2, 13.3]  # Temperaturen in C
durchschnittliche_temperatur = [8.1, 8.7, 8.7, 11.6, 18.8, 22.8, 25.4, 25.7, 21.4, 17.6, 12.6, 8.9]  # Temperaturen in C
minimale_temperatur          = [3, 3, 6.1, 8.2, 13.1, 16.8, 19, 19.3, 15.6, 12.3, 7.9, 4.4]  # Temperaturen in C

plt.figure(figsize=(14,9))

plt.plot(x, maximale_temperatur, label='Maximale Temperatur')
plt.plot(x, durchschnittliche_temperatur, label='Durchschnittliche Temperatur')
plt.plot(x, minimale_temperatur, label='Minimale Temperatur')

plt.legend(fontsize=15)

plt.xlabel('Monat', fontsize=25)
plt.ylabel('Temperatur T[C]', fontsize=25)
plt.title('Temperatur in Rom', fontsize=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.show()
```

Wir benutzen hier gleich wieder das Argument `fontsize` um die Gre des Textes anzupassen.

Als letztes wollen wir jetzt noch die Farben und den Linienstil anpassen. Das knnen wir auch ganz einfach ber die Argumente `color` fr die Farbe und `linestyle` fr den Linienstil regeln. Was steht uns aber zur Auswahl? Dazu schauen wir einfach in der Dokumentation nach:
- die Liste an verfgbaren Farben finder wir unter: https://matplotlib.org/stable/gallery/color/named_colors.html
- die Liste an verfgbaren Linienstilen finder wir unter: https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html.

Wir entscheiden uns fr die Durchschnittstemperatur eine schwarze durchgehende Linie (also `color='black', linestyle='-'`), fr die minimale Temperatur eine blaue gestrichelte Linie (also `color='blue', linestyle='--'`) und fr die maximale Temperatur eine gestrichelte rote Linie (also `color='red', linestyle='--'`) zu verwenden 

```python
x = range(1, 13) # Monate von 1 bis 12
maximale_temperatur          = [13.1, 14.3, 17.1, 19.2, 24.4, 28.7, 31.8, 32.1, 27.1, 22.8, 17.2, 13.3]  # Temperaturen in C
durchschnittliche_temperatur = [8.1, 8.7, 8.7, 11.6, 18.8, 22.8, 25.4, 25.7, 21.4, 17.6, 12.6, 8.9]  # Temperaturen in C
minimale_temperatur          = [3, 3, 6.1, 8.2, 13.1, 16.8, 19, 19.3, 15.6, 12.3, 7.9, 4.4]  # Temperaturen in C

plt.figure(figsize=(14,9))

plt.plot(x, maximale_temperatur, label='Maximale Temperatur', color='red', linestyle='--')
plt.plot(x, durchschnittliche_temperatur, label='Durchschnittliche Temperatur', color='black', linestyle='-')
plt.plot(x, minimale_temperatur, label='Minimale Temperatur', color='blue', linestyle='--')

plt.legend(fontsize=15)

plt.xlabel('Monat', fontsize=25)
plt.ylabel('Temperatur T[C]', fontsize=25)
plt.title('Temperatur in Rom', fontsize=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.show()
```

## <span style="color: #9C68C8;">Bonus: Gren verndern</span>


Um unsere Plots zugnglicher zu machen, knnen wir den Graphen auch noch ein bisschen vergrern. Dafr benutzen wir den Befehl `plt.figure()`. Dieser erstellt ein leeres Bild. Das Ganze knnt ihr euch wie eine unbemalte Leinwand oder ein unbedrucktes Blatt Papier vorstellen. Da wir natrlich die Leinwand oder das Papier vor dem Malen oder Drucken brauchen, muss dieser Befehl auch vor allen anderen stehen. Die Gre kann dann mit dem Argument `figsize` angepasst werden. Die Gre wird dabei durch `(B,H)` dargestellt, wobei `B` die Breite und `H` die Hhe sind. Matplotlib fllt dann die gesamte Leinwand bis zum Rand aus! So sieht das dann aus:



```python
plt.figure(figsize=(14,9)) # Dieser Befehl steht immer vor allen Befehlen die die Visualisierung betreffen

plt.plot()
plt.xlabel('x-Achsen Beschriftung')
plt.ylabel('y-Achsen Beschriftung')
plt.title('berschrift')

plt.show() # Dieser Befehl steht immer nach allen anderen Befehlen die Visualisierung betreffen
```

Das einzige Problem, das uns noch bleibt, ist, dass zwar das Bild grer ist, aber die Achsenbeschriftung nicht. Dies lsst sich durch das Argument `fontsize` (also Schriftgre auf Englisch) beheben. 

```python
plt.figure(figsize=(14,9)) 

plt.plot()
plt.xlabel('x-Achsen Beschriftung', fontsize=25)
plt.ylabel('y-Achsen Beschriftung', fontsize=25)
plt.title('berschrift', fontsize=30)

plt.show()
```

Dasselbe geht brigens auch mit den Zahlen an den Achsen in dem man den parameter `fontsize` in `plt.xticks()` und `plt.yticks()` anpasst

```python
plt.figure(figsize=(14,9)) 

plt.plot()
plt.xlabel('x-Achsen Beschriftung', fontsize=25)
plt.ylabel('y-Achsen Beschriftung', fontsize=25)
plt.title('berschrift', fontsize=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.show()
```
