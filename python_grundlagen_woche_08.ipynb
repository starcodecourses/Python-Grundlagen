## <span style="color: #9C68C8;">Text mit Python</span> 
### <span style="color: #9C68C8;">Text abspeichern</span> 

Das Speichern von Textdateien (Dateiendung `.txt`) kann man sich wie das Ablegen von Sachen in einem Fach eines Regals vorstellen. In dem Fall ist das Regal das Speichersystem des Computers, und das bestimmte Fach des Regals ist die Datei, an der wir arbeiten. Nach der Logik machen wir also 3 Schritte zum Abspeichern von Text:
1. Fach ffnen: `open()`
2. Daten ablegen: `write()`
3. Fach wieder schlieen: `close()`

In der Praxis sieht das dann so aus:

````python
datei = open("text.txt", "a")
datei.write("Lies mich!")
datei.close()
````

Jetzt taucht unsere kleine Datei `text.txt` links in der Dateibersicht auf. Auch hier kannst du draufklicken, um sie dir anzuschauen. Schreibe doch noch etwas dazu!

 <span style="color: #2E88B6;"> Unter der Haube macht der Befehl `plt.savefig()` auch nichts anderes, als die Datei anzulegen, die Datein hineinzuschreiben und die Datei dann wieder zu "schlieen". Das passiert allerdings alles automatisiert in diesem einen Befehl, sodass wir uns nicht darum kmmern mssen.</span>

### <span style="color: #9C68C8;">Text einlesen</span> 

````python
datei = open("text.txt", "r")
inhalt = datei.read()
print(inhalt)
datei.close()
````

Wenn du im Schritt davor noch etwas in die Textdatei dazugeschrieben hast, dann kannst du jetzt sehen, dass es hier auftaucht.

Das ist auch gut so: Das `a` steht fr "append" (englisch fr "anhngen" oder "dazuschreiben" in unserem Fall) und das `r` steht fr "read" (englisch fr "lesen", "einlesen").</span>

## <span style="color: #9C68C8;">Tabellen mit Python</span> 

Wir bleiben bei dem Beispiel mit den Wochentagen und erweitern die Daten nur um eine weitere Woche. Also so:


```python
temperaturen_wochentage = np.array([[25.5, 26.8, 24.3, 27.1, 23.9, 17.5, 18.5], [20.3, 19.3, 22.1, 23.5, 25.5, 24.9, 21.9]])
```

In welchem Format knnten sich die Daten noch speichern lassen? Wir haben bereits die Bilddatei kennengelernt sowie gerade die Textdatei. In der Praxis wrde man die Daten aber wahrscheinlich in einer Tabelle speichern. Das hat zwei Vorteile:

1. bessere bersichtlichkeit: Es ist eindeutig erkennbar, welchem Wochentag welche Temperatur zugeordnet ist. So, wie es in dem Code dargestellt ist, ist das nicht sofort ersichtlich.
2. bessere Verarbeitbarkeit: Tabllenformate sind stark standardisiert und funktionieren mit vielen unterschiedlichen Programmen. Deshalb braucht man sich zum Beispiel keine Sorgen zu machen, die Daten irgendwann mal von Hand abtippen zu mssen, weil sie nicht einlesbar sind.

Oft werden Tabellen as sogenannte *csv*-Dateien abgespeichert, also als Dateien mit der Endung `.csv`. Diese Dateien sind tatschlich einfach nur Textdateien, bei denen die Werte in einer Zeile durch ein Komma getrennt werden. Daher kommt auch der Name: *comma separated values - englisch fr Komma-getrennte Werte*. Das ganze sieht dann ungefhr so aus:

<center>

Nun wollen wir also die untenstehenden Daten als Tabelle abspeichern. Das Werkzeug dafr haben wir schon kennengelernt: NumPy eignet sich hervorragend dafr. Dessen Grundformate, die *NumPy-Arrays*, haben wir ja auch schon kennengelernt und sind also wieder mal unser Ausgangspunkt fr diese Aufgabe:

````python
temperaturen_wochentage = np.array([[25.5, 26.8, 24.3, 27.1, 23.9, 17.5, 18.5], [20.3, 19.3, 22.1, 23.5, 25.5, 24.9, 21.9]])
````

Die Funktion, die wir dafr verwenden, heit `.savetxt()`. Es gibt noch einige andere Methoden, die Tabellen zu speichern, die wren fr uns aber erstmal aufwendiger als `.savetxt`.

Wir bergeben der Funktion dann einfach die Informationen, die sie braucht:
1. Dateiname, in dem das Ganze abgespeichert werden soll: `"temperaturen.csv"`
2. Unser NumPy-Array: `temperaturen_wochentage`
3. Das Format, in dem die Zahlen geschrieben werden sollen: Wir wollen eine Gleitkommazahl (*float*) mit einer Nachkommastelle abspeichern. Das bedeutet fr den Computer: `"%.1f"`
4. Das Trennzeichen (*delimiter*), durch welches unsere Werte getrennt werden sollen. Typischerweise ist das ein Komma (*comma separated value*)

````python
np.savetxt("temperaturen.csv", temperaturen_wochentage, fmt="%.1f", delimiter=",")
````

Nach Ausfhren der oberen Zelle solltet ihr die Datei wieder links im Dateibrowser sehen.
